## Notebook to take the data generated by the parser and push it to Neo4j

```python
from py2neo import Graph, Node, Relationship
import os
from tqdm import tqdm
import pandas as pd
import json

NEO4J_USER = os.environ['NEO4J_USER']
NEO4J_PASSWORD = os.environ['NEO4J_PASS']
NEO4J_HOST = os.environ['NEO4J_HOST']


class NeoGraph:
    def __init__(self):
        self.g = Graph(host=NEO4J_HOST, user= NEO4J_USER, password = NEO4J_PASSWORD)

    def truncate(self):
        """Remove all nodes in the graph"""
        print("----- Truncating graph -----")
        tx = self.g.begin()
        result = tx.run('MATCH (n) DETACH DELETE n')
        tx.commit()
        return result

    def add_pieces(self, df):
        print("----- Starting Add companies process -----")
        tx = self.g.begin()
        for _, x in tqdm(df.iterrows(), total = len(df)):
            n = Node(x['piece_name'], name = x['piece'], piece_type = x['piece_type'], piece_name = x['piece_name'])
            tx.create(n)
        tx.commit()
        self.g.run("CREATE INDEX ON :Piece(name)")
        print("----- Adding pieces process complete -----")

    def create_links(self, df):
        print("----- Starting relationship creation process -----")
        for _, x in tqdm(df.iterrows(), total=df.shape[0]):
            cypher = f"MATCH (p1:{x['piece_name']} {{name:\'{x['piece']}\'}}),(p2:{x['capture_name']} {{name:\'{x['captures']}\'}}) CREATE (p1)-[:CAPTURED {{Pct_of_games : {x['pct_of_games']}}}]->(p2)"
            self.g.run(cypher)
        print("-----Relationship creation process complete -----")
    
    def run_query(self, query):
        res = n.g.run(query)
        return res.data()
```

```python
n = NeoGraph()
```

```python
# Load the captures_df file generated by the Parser Notebook
df = pd.read_pickle('../data/captures_df.pkl')
```

### Data Formatting and structuring

```python
df = df.assign(capture_type = lambda df: df['captures'].str.split('-').str.get(0))
```

```python
df = df.rename(columns={'game_id':'pct_of_games'})
```

```python
lookup = pd.DataFrame({'piece_type':df['piece_type'].unique(), 
              'piece_name':['Queen','Queen','Pawn','Bishop','Bishop','Pawn','Knight','Knight','Rook','Rook','King','King']})
```

```python
df = df.merge(lookup, on='piece_type').merge(lookup, left_on='capture_type', right_on='piece_type')
```

```python
df = df.rename(columns={'piece_name_x':'piece_name', 'piece_name_y':'capture_name','piece_type_x':'piece_type'})
```

### Insert to Neo4j

```python
# Delete existing data in database
n.truncate()
```

```python
n.add_pieces(df[['piece','piece_type','piece_name']].drop_duplicates())
```

```python
n.create_links(df)
```

### Create the projected graph to run the algorithms on

```python
# Create projected graph
query = """
CALL gds.graph.exists('projected-graph') YIELD exists;
"""
res = n.run_query(query)

if res[0].get('exists'):
    query = """
    CALL gds.graph.drop('projected-graph') YIELD graphName;
    """
    res = n.run_query(query)

query = """
CALL gds.graph.create('projected-graph','*', {
    CAPTURED: {
        type:'CAPTURED',
        properties:['Pct_of_games']
    }
})
YIELD graphName, nodeCount, relationshipCount;
"""
res = n.run_query(query)
```

### Algorithms
#### Pagerank

```python
query = """
CALL gds.pageRank.write('projected-graph', {
  maxIterations: 100,
  dampingFactor: 0.85,
  relationshipWeightProperty: 'Pct_of_games',
  writeProperty: 'pagerank'
})
YIELD nodePropertiesWritten AS writtenProperties, ranIterations, didConverge
"""
res = n.run_query(query)
print(res)
```

#### Weghted Degree Centrality

```python
query = """
CALL gds.alpha.degree.write('projected-graph',{
   relationshipWeightProperty: 'Pct_of_games',
   writeProperty: 'DegreeCentrality'
})
YIELD nodes, writeProperty
"""

res = n.run_query(query)
print(res)
```

### Getting data back from Neo4j and structuring for Visualisation
#### Formatting nodes

```python
res = n.run_query('match (a) return a')
```

```python
nodes = json.loads(json.dumps(res))
```

```python
nodes_plt = [x['a'] for x in nodes]
```

```python
for x in nodes_plt:
    x['id'] = x.pop('name')
    x['value'] = x.pop('DegreeCentrality')
    x['group']= x.pop('piece_name')
    x['label'] = x['id']
```

```python
nodes_out = nodes_plt.copy()
```

### Formatting edges

```python
res = n.run_query('match (source)-[link:CAPTURED]->(target) where link.Pct_of_games > 0.04 return source,link,target')
```

```python
links = json.loads(json.dumps(res))
```

```python
links_plt = []
```

```python
for link in links:
    links_plt.append({'from':link['source']['name'], 'to':link['target']['name'], 'value' : link['link']['Pct_of_games']})# 'label' : f"{link['link']['Pct_of_games']:.1%}"})
```

```python
out = {'nodes':nodes_out, 'edges':links_plt}
```

#### Writing data for the Degree Centrality measure

```python
with open('../data/data_degree_centrality.json', 'w') as outfile:
    json.dump(out, outfile)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Degree Centrality plot
plt.figure(figsize=(10,20))
ax = sns.barplot(data=pd.json_normalize(nodes_out).sort_values('value',ascending=False), y='label', hue='group', x='value', dodge=False,
                palette=["#97C2FC","#FB7E81", "#7AE142", "#EB7CF4","#FEFF00","#b434eb","#f8b1aa"])
ax.set_title('Weighted Degree Centrality')
ax.set_ylabel('Weighted Degree Centrality')
ax.set_xlabel('Piece')
plt.savefig('degree_centrality.png')
```

### Louvain Community Detection Algorithm

```python
query = """
CALL gds.louvain.stream('projected-graph', {includeIntermediateCommunities: true, relationshipWeightProperty: 'Pct_of_games'})

YIELD nodeId, communityId, intermediateCommunityIds
RETURN gds.util.asNode(nodeId).name AS name, communityId, intermediateCommunityIds
ORDER BY name ASC
"""
```

```python
res = n.run_query(query)
```

```python
louvain = pd.json_normalize(res).sort_values('communityId')
```

```python
louvain.head()
```

```python
# Add the Louvain value to the nodes so that the visualisation colours them by community
for node in out['nodes']:
    node['group'] = int(louvain[louvain['name'] == node['id']]['communityId'].values[0])
    _ = node.pop('value')
```

```python
# Write data for visualisation
with open('../data/data_louvain.json', 'w') as outfile:
    json.dump(out, outfile)
```
